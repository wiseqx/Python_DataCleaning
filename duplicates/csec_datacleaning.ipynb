## process

```python
df[['FIRST_NAME','MIDDLE_NAME']] = df["FIRST_NAME"].str.split(" ", 1, expand=True) #it separates first and middle name
```

```python
df.head(20)
```

```python
df['FIRST_NAME'] = df['FIRST_NAME'].map(lambda x: re.sub(r'[^a-zA-Z]', '', x))
```

```python
df['MIDDLE_NAME'] = df['MIDDLE_NAME'].map(lambda x: re.sub(r'[^a-zA-Z]', '', str(x)))
```

```python
df['LAST_NAME'] = df['LAST_NAME'].map(lambda x: re.sub(r'[^a-zA-Z]', '', x))
```

```python
df['DATE_OF_BIRTH'] = pd.to_datetime(df['DATE_OF_BIRTH'], errors='coerce')
df['DATE_OF_BIRTH'] = df["DATE_OF_BIRTH"].dt.strftime("%d%b%y")
```

```python
df.head(20)
```

#### create IDs

```python
df['IDENTIFIER1'] = df['FIRST_NAME'].astype(str) + df['LAST_NAME'].astype(str) + df['DATE_OF_BIRTH'].astype(str)
```

```python
df['IDENTIFIER2'] = df['FIRST_NAME'].astype(str) + df['MIDDLE_NAME'].astype(str) + df['LAST_NAME'].astype(str) + df['DATE_OF_BIRTH'].astype(str) + df['SEX'].astype(str)
```

```python
df['IDENTIFIER3'] = df['FIRST_NAME'].astype(str) + df['MIDDLE_NAME'].astype(str).str[0] + df['LAST_NAME'].astype(str) + df['DATE_OF_BIRTH'].astype(str)+ df['SEX'].astype(str)
```

#### handle school

```python
df['OVERALL_GRADE'].fillna(0, inplace=True)
```

```python
df.OVERALL_GRADE.replace({'VI': 1, 'V': 2, 'IV': 3, 'III': 4, 'II': 5, 'I':6,
                          'ABS':0, 'UNG+':0, 'UNG*':0, 'CANC':0, 'UNG-':0,
                          'WITH':0, 'WITHHELD':0, 'CANCELLED':0, 'L':0}, inplace=True)
```

```python
new_df = df.groupby(['FIRST_NAME', 'MIDDLE_NAME', 'LAST_NAME',
                     'DATE_OF_BIRTH', 'SEX', 'SUBJECT']).agg({'SCHOOL':'count', 'PRE_2014': 'max', 
                                                           'OVERALL_GRADE':'max','IDENTIFIER1':'first',
                                                          'IDENTIFIER2':'first','IDENTIFIER3':'first'}).reset_index()
```

```python
vals_to_replace = {2:'multiple', 3:'multiple', 4:'multiple', 5:'multiple', 6:'multiple', 1:1}
new_df['SCHOOL'] = new_df['SCHOOL'].map(vals_to_replace)
```
